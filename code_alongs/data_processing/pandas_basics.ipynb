```python
import pandas as pd
df = pd.read_csv("data/banking.csv")
df.head()
```

```python
df.astype({"loan": "float"})
```

```python
try:
    df.astype({"loan": "float"})
except ValueError as err:
    print(err)
```

```python
df[pd.to_numeric(df["loan"], errors="coerce").isna()]
```

```python
df[pd.to_numeric(df["loan"], errors="coerce").notna()]
df.shape
```

```python
df.head()
```

```python
try:
    df.astype({"loan": "int"})
except ValueError as err:
    print(err)
```

```python
df["loan"] = df["loan"].str.replace('no', "0").str.replace('yes', "1")
df.head()
```

```python
df.loan
```

```python
# convert loan str val to int
try:
    df.astype({"loan": "int"})
except ValueError as err:
    print(err)
```

```python
unk0wn_rows = df[df["loan"] == 'unk0wn']
```

```python
print(unk0wn_rows)
```

```python
# convert loan str val to int and unk0wn to 0 
df["loan"] = df["loan"].str.replace('unk0wn', "0") 
try:
    df.astype({"loan": "int"})
except ValueError as err:
    print(err)
```

```python
df = df[df["loan"].notna()]
df.shape
```

```python
df.info()
```

```python
loan_10_age = df.sort_values(by = "loan", ascending=False).iloc[:10]
loan_10_age[["age", "loan", "marital"]]
```

```python
df.plot(x="loan", y="loan_count", kind="scatter")
```
